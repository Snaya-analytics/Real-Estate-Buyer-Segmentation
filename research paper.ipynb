```python
# Real Estate Buyer Segmentation and Behavioral Analysis

## Abstract
This study analyzes 7305 buyers from real estate transactions. Data cleaning included ID normalization, sale price conversion, and removal of invalid rows. Exploratory Data Analysis and visualization (heatmap, bar chart, line chart, pie chart, boxplot) were applied. Findings reveal buyer segmentation patterns, loan behavior, satisfaction trends, and acquisition purposes.

## Introduction
Explain importance of real estate market, need for buyer segmentation, and objectives:
1. Identify buyer segments.
2. Analyze loan behavior.
3. Examine satisfaction trends.
4. Evaluate acquisition purposes.

## Literature Review
Summarize past studies on buyer segmentation, loan behavior, and satisfaction analysis.

## Data & Methodology
Dataset: buyers_final.csv (7305 rows).  
Columns: client_id, client_type, gender, country, region, acquisition_purpose, satisfaction_score, loan_applied, transaction_date, floor_area_sqft, sale_price.  
Cleaning: IDs normalized, sale_price numeric, transaction_date datetime, invalid rows dropped.  
Methods: EDA + visualization.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("buyers_final.csv")
df.head()
total_buyers = len(df)
avg_price = df['sale_price'].mean()
avg_satisfaction = df['satisfaction_score'].mean()
loan_pct = (df['loan_applied'].value_counts(normalize=True).get("Yes",0)*100)

print("Total Buyers:", total_buyers)
print("Average Sale Price:", avg_price)
print("Average Satisfaction:", avg_satisfaction)
print("Loan Applied %:", loan_pct)
sns.heatmap(df.groupby('client_type')[['sale_price','floor_area_sqft','satisfaction_score']].mean(), annot=True)
plt.show()
loan_chart = df.groupby(['client_type','loan_applied']).size().reset_index(name='count')
sns.barplot(x="client_type", y="count", hue="loan_applied", data=loan_chart)
plt.show()
trend = df.groupby('transaction_date')['satisfaction_score'].mean()
trend.plot(kind='line', title="Satisfaction Trend Over Time")
plt.show()
purpose_counts = df['acquisition_purpose'].value_counts()
purpose_counts.plot(kind='pie', autopct='%1.1f%%', title="Acquisition Purpose Distribution")
plt.show()
sns.boxplot(x="client_type", y="sale_price", data=df)
plt.show()
## Discussion
- Individual buyers show higher satisfaction but lower loan application rates.
- Corporate buyers contribute higher transaction values but moderate satisfaction.
- Investment dominates acquisition purposes.

## Conclusion & Recommendations
- Tailor loan products for individual buyers.
- Improve service quality for corporate buyers.
- Focus marketing on investment-driven buyers.
- Adjust seasonal strategies based on satisfaction trends.

## References
Add citations of past studies and data sources.
```
